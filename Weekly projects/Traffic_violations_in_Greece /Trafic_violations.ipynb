## This is my final dataframe

```python
#Now its long data format.Thank Nora.
final_df = pd.melt(df, id_vars=['ΕΙΔΟΣ_ΠΑΡΑΒΑΣΗΣ_'], value_vars=['2014', '2015', '2016', '2017', '2018'], value_name='number_of_tickets', var_name='year')
final_df
```

```python
final_df.number_of_tickets.sort_values(ascending=False)
```

```python
violations_allyears = final_df.groupby('ΕΙΔΟΣ_ΠΑΡΑΒΑΣΗΣ_').number_of_tickets.sum().sort_values(ascending=False)
violations_allyears
```

```python
violations_allyears.sort_values().plot(kind = 'barh', width=0.5, 
                                           figsize=(8, 10),
                                           title='Total number of violations for all years')
```

```python
#total number of violations per year
violations_per_year = final_df.groupby('year').number_of_tickets.sum().sort_values(ascending=False)
violations_per_year
```

```python
violations_per_year.plot(kind='pie', title='Number of violations per year')
```

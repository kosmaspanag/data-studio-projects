```python
#We did the assign think in order to get the sum of the two columns, then sort them and then
#the stack bar in oerder for the values to be displayed in the right order.
only_prisons_overload. \
assign(two = df['Συνολική_χωρητικότητα_σε_θέσεις']+ df['Κρατούμενοι']). \
sort_values('two', ascending=False)
```

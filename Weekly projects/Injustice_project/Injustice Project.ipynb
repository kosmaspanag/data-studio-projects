## Cleaning and reordering the data

```python
#full_name column & geting rid of extra spaces
df['full_name'] = df['First Name'] + " " + df['Last Name'] 
df
```

```python
df.full_name = df.full_name.replace('\s+', ' ', regex=True)
df.full_name
```

```python
# create years_of_conviction column
df['years_of_conviction'] = df['Exonerated'] - df['Convicted'] 
df
```

```python
df['years_of_conviction'].sum()
```

```python
#rename a few columns based on the coding manual

df = df.rename(index=str, columns={'*': 'Evidence_+_DNA',
                                   'FC': 'False_confession',
                                   'MWID': 'Mistaken_witness_ID',
                                   'F/MFE': 'False_misleading_forensic_evide',
                                   'P/FA': 'Perjury_false_accusation',
                                   'OM': 'Official_misconduct',
                                   'ILD': 'Inadequate_legal_defense',
                                   'Worst Crime Display':'Worst_Crime_Display',
                                   "List Add'l Crimes Recode" : "List_Add_Crimes_Recode"
                                  })
df.head()
```

```python
#I don't need this column
df = df.drop(columns=['Posting Date','ID'])
```
